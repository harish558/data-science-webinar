```python
import pandas as pd
import altair as alt
```

```python
demographic_data = pd.read_csv('demographic_data.csv')
```

```python
demographic_data.head()
```

### Starting with basic Visualization with Altair

#### Step 1 - Convert pandas data to altair data

```python
alt_data = alt.Chart(demographic_data)
```

#### Step 2 - Begin your visualization

```python
alt_data.mark_point().encode(
    x = alt.X('InternetUsers'),
    y = alt.Y('BirthRate')
)
```

```python
alt_data.mark_point().encode(
    x = alt.X('InternetUsers'),
    y = alt.Y('FertilityRate')
)
```

```python
alt_data.mark_point().encode(
    x = alt.X('BirthRate'),
    y = alt.Y('FertilityRate')
)
```

```python
alt_data.mark_boxplot(size=40).encode(
    x = alt.X('IncomeGroup'),
    y = alt.Y('InternetUsers'),
    tooltip = 'CountryName'
).properties(
    width=600,
    height=500
)
```

```python
alt_data.mark_boxplot(size=40).encode(
    x = alt.X('IncomeGroup'),
    y = alt.Y('BirthRate'),
    tooltip = 'CountryName'
).properties(
    width=600,
    height=500
)
```

```python
alt_data.mark_boxplot(size=40).encode(
    x = alt.X('IncomeGroup'),
    y = alt.Y('FertilityRate'),
    tooltip = 'CountryName'
).properties(
    width=600,
    height=500
)
```

```python
alt_data.mark_point().encode(
    x = alt.X('InternetUsers'),
    y = alt.Y('BirthRate'),
    color = 'IncomeGroup',
    tooltip = 'CountryName'
)
```
